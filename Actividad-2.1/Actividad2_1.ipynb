```python
#Carga de las librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga del archivo
df = pd.read_csv('Datos_Facturación.csv')
```

```python
#1.-CVE_CLPV      (valores: 1000 a 2000)

filtro1 = df[(df['CVE_CLPV'] >= 1000) & (df['CVE_CLPV'] <= 2000)]
print(filtro1)
```

```python
filtro1.to_csv('filtro1.csv')
```

```python
#.2-CVE_VEND     (Todas las claves excepto “5” y “4”)
filtro2 = df[~(df['CVE_VEND'] == 4 & (df['CVE_VEND'] == 5))]
print(filtro2)
```

```python
filtro2.to_csv('filtro2.csv')
```

```python
#.3-FECHA_ENT   (Fechas de “28/02/2022”)
filtro3 = df[df['FECHA_ENT'] == "2022-02-28 00:00:00"]
print(filtro3)
```

```python
filtro3.to_csv('filtro3.csv')
```

```python
filtro33 = df.groupby(['FECHA_ENT'])['FECHA_ENT'].count()
print(filtro33)
```

```python
pd.options.display.max_rows = 10
```

```python
#.4-CAN_TOT       (Cantidades menores a 5951.7)   o  STATUS (“E”)
filtro4 = df[(df['CAN_TOT'] < 5951.7) | (df['STATUS'] == "E")]
print(filtro4)
```

```python
filtro4.to_csv('filtro4.csv')
```

```python
#.5-Solo las columnas: CVE_DOC, FECHA_ENT, FECHA_VEN y CAN_TOT
filtro5 = df[['CVE_DOC', 'FECHA_ENT', 'FECHA_VEN', 'CAN_TOT']]
print(filtro5)
```

```python
filtro5.to_csv('filtro5.csv')
```

```python
#.6-Solo las filas de 7001-7099
filtro6 = df.iloc[7001 : 7100, :]
print(filtro6)
```
