```python
#Importación de librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
#Carga de la base
df = pd.read_csv("pais_hawaii.csv")
```

```python
df.columns
```

```python
df.head()
```

```python
#a) Los registros de 5 “host name” diferentes
#Agarramos los primeros 5 valores únicos
host_5 = df["host_nombre"].unique()[:5]
#Con isin mostramos solo los registros de los host en nuestra lista
filtroa = df[df["host_nombre"].isin(host_5)]
filtroa
```

```python
filtroa.to_csv("filtroa.csv")
```

```python
#b) Los registros de los host que se hallan unido a Airbnb “despues del año 2020”
#Revisamos que la fecha este en formato de fecha
df.info()
```

```python
#Pasamos la columna host_desde a formato de fecha
df["host_desde"] = pd.to_datetime(df["host_desde"])
```

```python
#Se define la fecha de inicio
fecha2020 = fecha_inicio = datetime(2020, 1, 1)
#Filtro b
filtrob = df[df["host_desde"] > fecha2020]
filtrob
```

```python
filtrob.to_csv("filtrob.csv")
```

```python
#c) Los registros de los host que responden “a más tardar en 1 día”
#Revisamos valores únicos para revisar escritura
df["tiempo_respuesta"].unique()
#Tomamos 'within a day' para hacer el filtro
filtroc = df[df["tiempo_respuesta"] == "within a day"]
filtroc
```

```python
filtroc.to_csv("filtroc.csv")
```

```python
#d) Los registros de las filas “Múltiplos de 200… (400, 600, 800, etc.)”
#Usamos el indice y "%" para que nos de sólo el número de filas que divididas entre 200 de como resultado 0
filtrod = df[df.index % 200 == 0]
filtrod
```

```python
filtrod.to_csv("filtrod.csv")
```

```python
#e) Los registros de los tipos de cuarto: “Entire home” y “Private room”
filtroe = df[(df["tipo_habitación"] == "Entire home") | (df["tipo_habitación"] == "Private room")]
filtroe
```

```python
filtroe.to_csv("filtroe.csv")
```

```python
#f) Los registros que “a lo mucho cuenten con 4 camas”
filtrof = df[df['camas'] <= 4]
filtrof
```

```python
filtrof.to_csv("filtrof.csv")
```

```python
#g) Los registros que “sean superhost”
filtrog = df[df['superhost'] == "t"]
filtrog
```

```python
filtrog.to_csv("filtrog.csv")
```

```python
#h) Los registros de las columnas “impares”
#Desde la columna 1 se toman cada 2 columnas, quedando solo impares
filtroh = df.iloc[:,1::2]
filtroh
```

```python
filtroh.to_csv("filtroh.csv")
```
